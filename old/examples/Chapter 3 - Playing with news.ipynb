```python
articles = json.load( open('../yle.json') ) ## note, file path changed as we're in a subfolder

counter_hallitus = 0
counter_oppositio = 0

for article in articles:
    
    subtitle = article['subtitle'].lower()
    
    if 'hallitus' in subtitle:
        counter_hallitus = counter_hallitus + 1
        
    if 'oppositio' in subtitle:
        counter_oppositio = counter_oppositio + 1
        
if counter_hallitus > counter_oppositio:
    print 'Hallitus on puhutumpi'
    
if counter_hallitus < counter_oppositio:
    print 'Oppositio on puhutumpi'
```
