`twitter_network.txt` is a file that shows the mention network under spesific hashtag in Twitter. The file looks like

`2933312565 -> 3152342314
434383089 -> 1240030489
1116331430 -> -`

i.e. it is in format of `who -> who is mentioned` where `-` means no user was mentioned.

**Playing around**

* Who is the most active tweeter (sends out most tweets)?
* Who receives most tweets?
* Who sends out most tweets which do not mention others?

```python
for line in open('twitter_network.txt'):
    
    line = line.strip() ## remove extra spaces in the begin and end of line
    _from = line.split(' -> ')[0] ## part before ->
    _to = line.split(' -> ')[1] ## part after ->
    ## you should not touch these lines ^^^
    
    print _from, _to
```
